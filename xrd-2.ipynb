```python
from pymatgen.ext.matproj import MPRester
from pymatgen import Structure
from pymatgen.analysis.diffraction import xrd
import matplotlib.pyplot as plt
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer

with MPRester("KCN5RO8FHahQJZKZ") as m:

    LiZnP = m.get_structure_by_material_id("mp-10182")
    
LiZnP.to(filename="LiZnP.cif")
struct_LiZnP = SpacegroupAnalyzer(LiZnP)  
struct_conv = struct_LiZnP.get_conventional_standard_structure()
```

```python
calculator = xrd.XRDCalculator(1.54060)
```

```python
infos = calculator.get_pattern(struct_conv)
print("1er pic : Intensité =",infos.y[0])
print("Indices hkl associés =", infos.hkls[0])
print("")
print("2ème pic : Intensité =",infos.y[1])
print("Indices hkl associés =", infos.hkls[1])
print("")
print("3ème pic : Intensité =",infos.y[2])
print("Indices hkl associés =", infos.hkls[2])
```

```python
calculator.show_plot(struct_conv, annotate_peaks = True)
```
